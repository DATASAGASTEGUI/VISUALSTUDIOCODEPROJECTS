<span style="color:blue; font-size:15px">QUE ES OPENCV</span>

<div class=text-justify>
<ol>

<li>Es una biblioteca libre de visión artificial, originalmente desarrollada por Intel. Su primera versión alfa fue en el mes de enero de 1999.</li><br>

<li>Se ha utilizado en aplicaciones desde sistemas de seguridad con detección de movimiento, hasta control de procesos donde se requiere reconocimiento de objetos.</li><br>

<li>Su publicación se da bajo licencia BSD, la cual permite que sea usada para propósitos comerciales y de investigación libre.</li><br>

<li>Es multiplataforma, existiendo versiones para GNU/Linux, Mac OS X, Windows y Android.</li><br>

<li>Contiene más de 500 funciones que abarcan una gran gama de áreas en el proceso de visión, tales como reconocimiento de objetos, visión estéreo, maching learning, procesamiento de imagenes etc.</li><br>

<li>Contiene diferentes interfaces para diferentes lenguajes de programación como c, c++, python, java.</li>
    
</ol>
</div>

<span style="color:blue; font-size:15px">CODIGO OPENCV</span>

<div class=text-justify>
<ol>
<li>Importar la libreria opencv</li>
import cv2

<li>Ruta realtiva o absoluta del archivo que contiene la imagen</li>
img = cv2.imread('imagen/letra_color.bmp')

<li>Carga una imagen en color. Se descuidará cualquier transparencia de la imagen. Es la bandera predeterminada.</li>
cv2.IMREAD_COLOR
    
<li>Carga la imagen en modo escala de grises.</li>
cv2.IMREAD_GRAYSCALE
    
<li>Carga la imagen como tal, incluido el canal alfa.</li>  
cv2.IMREAD_UNCHANGED: 

<li>'Imagen Color' es el nombre de la ventana donde se muestra la imagen, img es la imagen que se mostrará en la ventana.</li>
cv2.imshow('Imagen Color',img)

<li>Espera hasta que se presiona una tecla</li>
cv2.waitKey(0) 

<li>Destruye la ventana que muestra la imagen</li>
cv2.destroyAllWindows()
</ol>
</div><br>

EJEMPLO 01

```python
import cv2

img = cv2.imread('imagen/letra_blanco_negro.bmp', cv2.IMREAD_UNCHANGED) 
cv2.imshow('Imagen BlancoNegro',img) 
print(img)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread('imagen/letra_gris.bmp', cv2.IMREAD_GRAYSCALE) 
cv2.imshow('Imagen Gris',img) 
print(img)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread('imagen/letra_color.bmp', cv2.IMREAD_COLOR) 

[f,c,n] = img.shape
imageColorR = img[:,:,2]
imageColorG = img[:,:,1]
imageColorB = img[:,:,0]

print('R: ', imageColorR)
print('G: ', imageColorG)
print('B: ', imageColorB)

cv2.imshow('Imagen Color',img) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
import cv2

img_c = cv2.imread('imagen/color.jpg', cv2.IMREAD_COLOR)
img_g = cv2.imread('imagen/gris.jpg', cv2.IMREAD_GRAYSCALE) 
img_b = cv2.imread('imagen/blanco_negro.jpg', cv2.IMREAD_COLOR) 

cv2.namedWindow('Color', cv2.WINDOW_AUTOSIZE)
cv2.namedWindow('Gris', cv2.WINDOW_AUTOSIZE)
cv2.namedWindow('BlancoNegro', cv2.WINDOW_AUTOSIZE)

cv2.imshow('Color',img_c) 
cv2.imshow('Gris',img_g) 
cv2.imshow('BlancoNegro',img_b) 

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image = cv2.imread('imagen/pinguino.jpg', cv2.IMREAD_COLOR) 

[f,c,n] = img.shape
imageColorR = np.copy(image)
imageColorG = np.copy(image)
imageColorB = np.copy(image)

imageColorR[:,:,1] = 0
imageColorR[:,:,0] = 0

imageColorG[:,:,2] = 0
imageColorG[:,:,0] = 0

imageColorB[:,:,2] = 0
imageColorB[:,:,1] = 0

cv2.namedWindow('R', cv2.WINDOW_AUTOSIZE)
cv2.namedWindow('G', cv2.WINDOW_AUTOSIZE)
cv2.namedWindow('B', cv2.WINDOW_AUTOSIZE)

cv2.imshow('R',imageColorR) 
cv2.imshow('G',imageColorG)
cv2.imshow('B',imageColorB)

cv2.waitKey(0) 
cv2.destroyAllWindows()

cv2.imwrite('imagen/pinguino_r.jpg',imageColorR)
cv2.imwrite('imagen/pinguino_g.jpg',imageColorG)
cv2.imwrite('imagen/pinguino_b.jpg',imageColorB)
```
